```python
import pandas as pd
```

```python
beyonce = pd.read_csv("beyonce_songs.csv")
bruno_mars = pd.read_csv("bruno_mars_songs.csv")
ed_sheeran = pd.read_csv("ed_sheeran_songs.csv")
ice_cube = pd.read_csv("ice_cube_songs.csv")
maluma = pd.read_csv("maluma_songs.csv")
taylor_swift = pd.read_csv("taylor_swift_songs.csv")
```

```python
full_df = pd.concat([beyonce, ed_sheeran, ice_cube, maluma, taylor_swift], ignore_index=True)
full_df
```

```python
import numpy as np
full_df.isnull().sum()
```

```python
full_df = full_df.drop(columns = "Unnamed: 0") # Drops Unnamed column
full_df.to_csv("full_df.csv")
full_df
```

```python
grouped = full_df.groupby('artist_name')

averages = grouped.mean()
print(averages)
```

```python
averages_df = pd.DataFrame(averages)
averages_df
```

```python
averages_df.info()
```

```python
averages_df = averages_df.drop(columns = "track_number")
names = ["Beyonce", "Ed Sheeran", "Ice Cube", "Maluma", "Taylor Swift"]
averages_df['artist_name'] = names
averages_df
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
artists = averages_df['artist_name']
popularity = averages_df['popularity']

acousticness = averages_df['acousticness']
danceability = averages_df['danceability']
energy = averages_df['energy']
instrumentalness = averages_df['instrumentalness']
liveness = averages_df['liveness']
loudness = averages_df['loudness']
speechiness = averages_df['speechiness']
tempo = averages_df['tempo']
valence = averages_df['valence']
```

```python
plt.bar(artists, popularity)
plt.xlabel("Artist")
plt.ylabel("Popularity Score")
plt.title("Average Song Popularity by Artist")
```

```python
N = 5

ind = np.arange(N)

width = 0.2
plt.figure(figsize=(10,7))

plt.bar(artists, acousticness, width, label = 'Acousticness')
plt.bar(ind + 0.2, energy, width, label = 'Energy')
plt.bar(ind + 0.4, danceability, width, label = 'Danceability')
plt.bar(ind + 0.6, valence, width, label = 'Valence')
plt.bar(ind + 0.8, speechiness, width, label = 'Speechiness')
plt.legend(bbox_to_anchor = (1, 1))
plt.xticks(ind + width / 0.5)
plt.xlabel("Artist Name")
plt.ylabel("Attribute Score")

plt.title("Song Attributes Scores by Artist")
```
