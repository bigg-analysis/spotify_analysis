```python
import pandas as pd
song_df = pd.read_csv("maluma_songs.csv")
song_df
```

```python
new_song_df = song_df.iloc[:,9:18]
new_song_df
```

```python
new_song_df.isnull().sum()
```

```python
x = new_song_df.drop(columns = 'popularity')
x
```

```python
y = new_song_df['popularity']
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x,y)
```

```python
r_squared = lr.score(x,y)
print(r_squared)
```
